```python
#importing necessary libraries


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset

df=pd.read_csv('world_population.csv')
df
```

<h1> Preprocessing </h1>

```python
#getting an info about various columns.

df.info()
```

```python
#renaming a few columns

df=df.rename(columns={f'Area (km\N{SUPERSCRIPT TWO})':'Area (sq km)',f'Density (per km\N{SUPERSCRIPT TWO})':'Density (sq km)'})
```

```python
#checking columns

df.columns
```

```python
#plotting countries based on population as of 2022

df_top5=df.sort_values(by='2022 Population').tail(5)
df_low5=df.sort_values(by='2022 Population').head(5)



plt.figure(figsize=(8,8))
sns.barplot(x='Country',y='2022 Population',data=df_top5)
plt.title('Countries having highest Population in 2022')
plt.ylabel('Population (in billion)')
plt.show()

plt.figure(figsize=(8,8))
sns.barplot(x='Country',y='2022 Population',data=df_low5)
plt.title('Countries having lowest Population in 2022')
plt.ylabel('Population')
plt.show()
```

It is visible that India and China are the most Populous Countries. <br><br>
Whereas Vatican city is the Country with lowest population.

```python
#plotting countries based on their area

df_top5=df.sort_values(by='Area (sq km)').tail(5)
df_low5=df.sort_values(by='Area (sq km)').head(5)

plt.figure(figsize=(8,8))
sns.barplot(x='Country',y='Area (sq km)',data=df_top5)
plt.title('Countries wih most Area')
plt.ylabel('Area (in  10 million sq km)')
plt.show()

plt.figure(figsize=(8,8))
sns.barplot(x='Country',y='Area (sq km)',data=df_low5)
plt.title('Countries with least Area')
plt.ylabel('Area (in sq km)')
plt.show()
```

It can be seen that russia is the largest country in terms of its area. <br><br>
Vatican city is the smallest country in terms of its area.

```python
#plotting countries based on their population density 

df_top5=df.sort_values(by='Density (sq km)').tail(5)
df_low5=df.sort_values(by='Density (sq km)').head(5)

plt.figure(figsize=(8,8))
sns.barplot(x='Country',y='Density (sq km)',data=df_top5)
plt.title('Countries having highest population density')
plt.ylabel('Population Per Square km')
plt.show()

plt.figure(figsize=(8,8))
sns.barplot(x='Country',y='Density (sq km)',data=df_low5)
plt.title('Countries having lowest population density')
plt.ylabel('Population Per Square km')
plt.show()
```

It can be seen that Macau and Monaco are the countries with largest population density. <br><br>
Greenland and Falkland Islands have lowest density of population.

```python
#plotting countries based on their population growth rate

df_top5=df.sort_values(by='Growth Rate').tail(5)
df_low5=df.sort_values(by='Growth Rate').head(5)

plt.figure(figsize=(8,8))
sns.lineplot(x='Country',y='Growth Rate',data=df_top5)
plt.title('Countries having highest population growth rate')
plt.ylabel('growth rate')
plt.grid()
plt.show()

plt.figure(figsize=(8,8))
sns.lineplot(x='Country',y='Growth Rate',data=df_low5)
plt.title('Countries having lowest population growth rate')
plt.ylabel('growth rate')
plt.grid()
plt.show()

```

We can see that Moldova has the highest population growth rate. <br><br>
Ukraine has the lowest population growth rate.
